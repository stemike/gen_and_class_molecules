```python
import pandas as pd
import matplotlib.pyplot as plt
import json
import math
import numpy as np
import seaborn as sns
```

```python
#Plot ReLu
from matplotlib import pyplot
 
# rectified linear function
def rectified(x):
	return max(0.0, x)
 
# define a series of inputs
series_in = [x for x in range(-10, 11)]
# calculate outputs for our inputs
series_out = [rectified(x) for x in series_in]
# line plot of raw inputs to rectified outputs
pyplot.plot(series_in, series_out)
pyplot.title("Rectified Linear Unit (ReLU)")
pyplot.xticks([-10, -5, 0, 5, 10], [-10, -5, 0, 5, 10])
pyplot.show()
```

```python
df = pd.read_csv("output/labeledMolecules.csv")
```

```python
non_toxic = df[df.toxicity == 0].smiles
toxic = df[df.toxicity == 1].smiles
#non_toxic.to_csv("output/safe_smiles.txt",header = False, index = False)
print(df.shape[0])
print(non_toxic.size)
print(non_toxic.size * 100 / df.shape[0])
```

```python
non_toxic = non_toxic.rename("Length distribution of non toxic molecules")
sns.distplot(non_toxic.str.len(), hist=True, kde=True, 
             bins=int(180/5), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
```

```python
labels = 'Toxic', 'Non-Toxic'
sizes = [toxic.size, non_toxic.size]
colors = ['orangered', 'darkgreen']
explode = (0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

plt.title("Toxic vs non-toxic compounds")
plt.show()
```
